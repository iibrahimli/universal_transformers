```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
results_dir = Path("../results/wmt14/")

tr_loss = pd.read_csv(results_dir / "tr_loss.csv")
val_loss = pd.read_csv(results_dir / "val_loss.csv")
val_bleu = pd.read_csv(results_dir / "val_bleu.csv")
val_bertscore = pd.read_csv(results_dir / "val_bertscore.csv")
```

## Training and validation loss

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 3))

# loss curves
ax[0].plot(tr_loss["Step"], tr_loss["zesty-flower-29 - tr.loss"], label="Train")
ax[0].plot(val_loss["Step"], val_loss["zesty-flower-29 - val.loss"], linewidth=3, label="Validation")
ax[0].set_xlabel("Step (million)")
ax[0].set_ylabel("Loss")
ax[0].legend()

# bleu
ax[1].plot(val_bleu["Step"], val_bleu["zesty-flower-29 - val.bleu"], label="BLEU")
ax[1].set_xlabel("Step (million)")
ax[1].set_ylabel("BLEU")
ax[1].legend()

# bertscore
ax[2].plot(val_bertscore["Step"], val_bertscore["zesty-flower-29 - val.bertscore"], label="BERT-score")
# ax[2].set_xlim(0)
ax[2].set_xlabel("Step (million)")
ax[2].set_ylabel("BERT-score")
ax[2].legend()

plt.tight_layout()
plt.savefig(results_dir / "wmt14_training.png", dpi=300)
plt.show()
```
